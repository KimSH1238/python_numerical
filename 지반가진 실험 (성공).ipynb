```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as so

x1=np.array([285, 270, 295, 260, 300])
r1=x1/285

y1=np.array([33, 15, 15, 10, 13])
R1=y1/4

def f(x,a):
    A=(2*a*x)**2
    B=(1-x**2)**2
    return ((1+A)/(A+B))**0.5

popt1, pcov1=so.curve_fit(f,r1,R1)
rr1=np.linspace(min(r1),max(r1),1000)
plt.plot(rr1,f(rr1,popt1))
plt.plot(r1,R1,'ro')

#커브피팅 하지 않고 보간으로 하더라도 더 나아보인다.
```
