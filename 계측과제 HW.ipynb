```python
import numpy as np
import matplotlib.pyplot as plt

a=np.array([51.9, 51., 50.3, 49.6, 51., 50., 48.9, 50.5,
            50.9, 52.4, 51.3, 50.7, 52., 49.4, 49.7, 50.5, 50.7, 49.4, 49.9, 49.2])
b=np.array([51.9, 48.7, 51.1, 51.7, 49.9, 48.8, 52.5, 51.7, 51.3, 52.6, 49.4,
            50.3, 50.3, 50.2, 50.9, 52.1, 49.3, 50.7, 50.5, 49.7])
c=np.array([51.1, 50.1, 51.4, 50.5, 49.7, 51.6, 51., 49.5, 52.4,
            49.5, 51.6, 49.4, 50.8, 50.8, 50.2, 50.1, 52.3, 48.9, 50.4, 51.5])
# b[19]에서 에러 발생

print("첫번째 set")
print("평균은 ",np.mean(a))
print("분산은 ",np.var(a))
print("표준 편차는",np.std(a),"\n")

print("두번째 set")
print("평균은 ",np.mean(b))
print("분산은 ",np.var(b))
print("표준 편차는",np.std(b),"\n")

print("세번째 set")
print("평균은 ",np.mean(c))
print("분산은 ",np.var(c))
print("표준 편차는",np.std(c),"\n")
```
