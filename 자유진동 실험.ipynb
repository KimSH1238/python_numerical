```python
import numpy as np

def nl_fit(x,y0):
    n=len(x)
    y=np.log(y0)
    xsum=sum(x)
    ysum=sum(y)
    xmean=np.mean(x)
    ymean=np.mean(y)
    xysum=sum(x*y)
    x2sum=sum(x**2)
    A=(n*xysum-xsum*ysum)/(n*x2sum-xsum**2)
    B=ymean-A*xmean
    
    print('A = %8.4f B = %8.4f' % (A,B))
    b=np.exp(B)
    a=A
    return a,b

def f(a,b,x):
    return b*np.exp(a*x)

m=np.array([0.5504, 0.7401, 0.9345])
wd=np.array([20.3951, 17.5880, 15.6523])
a,b=nl_fit(m,wd)

print('a = %8.4f b = %8.4f' % (a,b))
print('wd = %8.4f exp(%8.4f m)' % (b,a))
```
