```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x*np.cos(2*x)
def fd(x):
    return np.cos(2*x)-2*x*np.sin(2*x)

def fwd(xi):
    xpi=xi+h
    fdf=(f(xpi)-f(xi))/h
    return fdf
def bd(xi):
    xmi=xi-h
    bdf=(f(xi)-f(xmi))/h
    return bdf
def cd(xi):
    xpi=xi+h
    xmi=xi-h
    cdf=(f(xpi)-f(xmi))/(2*h)
    return cdf

h=1.
xi=np.linspace(2,4,3)

fw=list()
b=list()
c=list()
ff=list()

for i in xi:
    fw.append(fwd(i))
#    print(f'forward diff: {fwd(i):0.4f}')
    b.append(bd(i))
#    print(f'backward diff: {bd(i):0.4f}')
    c.append(cd(i))
#    print(f'centered diff: {cd(i):0.4f}')
    ff.append(fd(i))
#    print(f'real value: {fd(i):0.4f}\n')
print(fw)
print(b)
print(c)
print(ff)

xi=np.linspace(2,4,500)

fw=list()
b=list()
c=list()
ff=list()

for i in xi:
    fw.append(fwd(i))
#    print(f'forward diff: {fwd(i):0.4f}')
    b.append(bd(i))
#    print(f'backward diff: {bd(i):0.4f}')
    c.append(cd(i))
#    print(f'centered diff: {cd(i):0.4f}')
    ff.append(fd(i))
#    print(f'real value: {fd(i):0.4f}\n')

#graph

fwax= plt.plot(fw, label='forward')
bax= plt.plot(b, label='backward')
cax= plt.plot(c, label='centered')
ffax= plt.plot(ff, label='real function')
plt.legend()
#ax.legend([fwax,bax,cax,ffax], ['forward','backward','centered','real function'])
```

오차가 예상과 다르고 세가지 모두 크게 작용한다는 것을 인지할 수 있다. 이를 극복하기 위해서는 h값을 줄여야 한다. 아래에서 동일하게 h값을 줄여보았다.
